```python
primary=primary[primary['Indicator']=='Rate of out-of-school children of primary school age, female (%)']
```

```python
primary.drop(['EDULIT_IND','TIME','Flag Codes','Flags','Indicator'],axis=1,inplace=True)
```

```python
primary.rename(columns={'Value':'primary_females_pct_out'},inplace=True)
```

```python
mortality_df['Year']=mortality_df['Year'].astype(int)
```

```python
full_set=primary.merge(mortality_df,left_on=['LOCATION','Time'],right_on=['ISO Code','Year'])
```

```python
rate_set=primary.merge(mortality_rate_df,left_on=['LOCATION','Time'],right_on=['ISO Code','Year'])
```

```python
rate_set.drop(['LOCATION','Country','Time'],axis=1,inplace=True)
```

```python
rate_set.dropna(subset=['primary_females_pct_out', 'mort_rate_U5'])
```

```python
rate_set_2016=rate_set[rate_set['Year']==2016]
```

```python
rate_set_2016
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model
```

```python
X = rate_set_2016[['primary_females_pct_out']]
y = rate_set_2016['mort_rate_U5'].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)
```
