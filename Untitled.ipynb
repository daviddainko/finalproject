```python
import pandas as pd
import numpy as np
```

```python
gdp_per_capita=pd.read_csv("Inputs/GDP_Per_Capita.csv",header=4)
```

```python
out_of_school_females=pd.read_csv("Inputs/EDULIT_DS_28042019225423638.csv")
```

```python
child_mortality=pd.read_excel("Inputs/U5MR_mortality_rate_2018.xlsx",sheet_name=0,header=10)
```

```python
# list(child_mortalic
mortality_2016_df=child_mortality[['ISO Code', 2016.5]].dropna()
```

```python
out_of_school_females['Indicator'].unique()
```

```python
female_ed=out_of_school_females[out_of_school_females['Indicator']=='Rate of out-of-school children, adolescents and youth of primary and secondary school age, female (%)']
```

```python
female_ed=female_ed[female_ed['Time']==2016]
```

```python
female_ed=female_ed[['LOCATION','Time','Value']]
```

```python
female_ed.rename(columns={'Value':'% females out of school'},inplace=True)
```

```python
female_ed.head()
```

```python
mortality_2016_df.rename(columns={2016.5:'child_mortality_rate'},inplace=True)
```

```python
mortality_2016_df.head()
```

```python
gdp_2016_df=gdp_per_capita[['Country Code','2016']].dropna()
```

```python
gdp_2016_df.rename(columns={'2016':'GDP_Per_Capita'},inplace=True)
```

```python
gdp_2016_df.head()
```

```python
total_df=pd.merge(female_ed,mortality_2016_df,left_on='LOCATION',right_on='ISO Code')
```

```python
total_df=pd.merge(total_df,gdp_2016_df,left_on='LOCATION',right_on='Country Code')
```

```python
total_df=total_df[['Country Code','% females out of school','GDP_Per_Capita','child_mortality_rate']]
```

```python
total_df.head()
```

```python
X = total_df[['% females out of school','GDP_Per_Capita']]
y = total_df['child_mortality_rate'].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)

### END SOLUTION 

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```

```python
model.get_params()
```

```python
print('Coefficients: \n', model.coef_)
```

```python
model.coef_[0][0]
```
